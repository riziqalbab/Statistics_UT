### 1. Penumpang ke stasiun

- a Ruang sample
    Kemungkinan hasil yang dapat terjadi adalah:

    S = {tepat waktu, terlambat}

- b Kejadian (A)

    Misalnya kejadian yang ingin dihitung peluangnya adalah penumpang tepat waktu.

    A = {tepat waktu}

- c Saat seseorang membeli makanan lewat aplikasi ojek online, ada ketidakpastian apakah makanan akan sampai cepat atau terlambat.

    Ruang sampelnya:

    S = {cepat, terlambat}

    Misalnya kejadian A adalah makanan sampai cepat, maka:

    A = {cepat}

    Peluang digunakan untuk memperkirakan apakah kita perlu memesan lebih awal atau bisa menunggu waktu makan.

## 2. Pengiriman 3 Paket

Setiap paket dapat **Tepat waktu (T)** atau **Terlambat (L)**.

Ruang sampel:

$$
S = \{TTT, TTL, TLT, TLL, LTT, LTL, LLT, LLL\}
$$

Karena semua kemungkinan memiliki peluang yang sama, maka peluang setiap hasil adalah:

$$
P = \frac{1}{8}
$$

Kejadian **A** adalah paling sedikit 2 paket tepat waktu.

$$
A = \{TTT, TTL, TLT, LTT\}
$$

Kejadian **B** adalah paket ke-1 terlambat dan paket ke-2 tepat waktu.

$$
B = \{LTT, LTL\}
$$

### a) Menghitung $P(A \cap B)$

Irisan antara kejadian **A** dan **B** adalah:

$$
A \cap B = \{LTT\}
$$

Maka:

$$
P(A \cap B) = \frac{1}{8}
$$

### d) Menghitung $P(A \mid B)$

Rumus peluang bersyarat:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

Diketahui:

$$
P(A \cap B) = \frac{1}{8}
$$

$$
P(B) = \frac{2}{8}
$$

Maka:

$$
P(A \mid B) = \frac{\frac{1}{8}}{\frac{2}{8}}
$$

$$
P(A \mid B) = \frac{1}{2}
$$

Jadi, hasil akhirnya adalah:

$$
P(A \cap B) = \frac{1}{8}
$$

$$
P(A \mid B) = \frac{1}{2}
$$
