## PERTANYAAN 1

```r
## MEMBUAT VARIABLE M UNTUK MENYIMPAN MATRIKS

M <- matrix(c(
  5, 7, 5, 4,
  2, 9, 7, 5,
  3, 3, 4, 2,
  8, 1, 0, 3
), nrow = 4, byrow = TRUE)

M
```

```r
print(M)
```

```r
# # a. Perkalian elemen demi elemen
a <- M * M

print(a)
```

```r
# b. Perkalian matriks M dengan M
b <- M %*% M

print(b)
```

```r
# c. Invers matriks M

c <- solve(M)
print(c)
```

```r
# d. Transpos matriks M
t(M)
```

```r

# e. MT × M
t(M) %*% M
```

## PERTANYAAN 2

```r
# Matriks M
M <- matrix(c(
  5, 7, 5, 4,
  2, 9, 7, 5,
  3, 3, 4, 2,
  8, 1, 0, 3
), nrow = 4, byrow = TRUE)

# Vektor a
a <- c(1, 4, 5, 6)

# Matriks B
B <- matrix(c(
  1, 2, 4, 4,
  1, 2, 3, 4
), nrow = 2, byrow = TRUE)
```

```r
# Mengambil anak matriks M1

M1 <- matrix(c(M[2,2], M[2,4], 
              M[3,4], M[4,2]), 
            nrow = 2, ncol = 2, byrow = TRUE)

M1
```

```r
# Mengambil anak matriks M2

M2 <- matrix(c(M[1,2], M[1,4], 
              M[3,3], M[4,4]), 
            nrow = 2, ncol = 2, byrow = TRUE)

print(M2)
```

```r
# Menambahkan a ke baris ke-5 dari M

M3 <- rbind(M, a)
M3
```

```r
# d. Menambahkan Bᵀ ke kolom ke-1 dan 2 dari M

M4 <- cbind(t(B), M)
M4
```

## PERTANYAAN 3

### PERKALIAN MATRIX

```r
A <- matrix(c(
  1, 2,
  3, 4
), nrow = 2, byrow = TRUE)

B <- matrix(c(
  5, 6,
  7, 8
), nrow = 2, byrow = TRUE)

C <- A %*% B
C
```

### TRANSPOS MATRIX

```r
M <- matrix(c(
  2, 1,
  5, 3
), nrow = 2, byrow = TRUE)

# Transpos
t(M)

# Invers
solve(M)
```
