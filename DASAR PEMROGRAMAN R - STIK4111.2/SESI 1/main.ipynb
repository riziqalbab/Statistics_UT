## Syarat penamaan variabel dalam bahasa pemrograman R

- Nama variabel harus dimulai dengan huruf atau tanda garis bawah (_), dan tidak boleh dimulai dengan angka.
- Nama variabel hanya boleh mengandung huruf, angka, dan tanda garis bawah (_), dan tidak boleh mengandung spasi atau karakter khusus lainnya.
- Nama variabel tidak boleh menggunakan kata kunci (reserved words) yang sudah digunakan dalam bahasa pemrograman R, seperti "if", "else", "for", "while", dll
